### Data Cleaning

Using Copy of Dataset for Cleaning

```python
df_copy=df.copy()
```

```python


df_copy['Reviews'].unique()
```

#### Better Solution to handle Astype Conversion

```python
try:
    df_copy['Reviews']=df_copy['Reviews'].astype(int)
except ValueError:
    mask = df_copy['Reviews'].apply(lambda x: not str(x).isdigit())  #imp: creating mask for keeping str
    df_copy = df_copy[~mask]
    df_copy['Reviews']=df_copy['Reviews'].astype(int)  ## Convert Review Datatype to int
```

```python
## Convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000000')
df_copy['Size']=df_copy['Size'].str.replace('k','000')
df_copy.head()
```

```python
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy
```

```python
df_copy['Size']=df_copy['Size'].astype(float)
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
#combine  removal
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']

for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
try:
    df_copy['Installs']=df_copy['Reviews'].astype(int)
    df_copy['Price']=df_copy['Price'].astype(float)
except ValueError:
    mask1 = df_copy['Installs'].apply(lambda x: not str(x).isdigit())  #imp: creating mask for keeping str
    mask2 = df_copy['Price'].apply(lambda x: not str(x).isdigit())
    df_copy = df_copy[~mask1]
    df_copy = df_copy[~mask2]
    df_copy['Installs']=df_copy['Reviews'].astype(int)
    df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
# last updated
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('Data/google_data_cleaned.csv')
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store


## What are the Top 5 most installed Apps in Each popular Categories ??

```python
most_insalled_apps=df_copy.groupby(['Category','App'])['Installs'].sum().sort_values(ascending=False).reset_index()
most_insalled_apps
```

```python
apps=['GAME','COMMUNICATION','PRODUCTIVITY','SOCIAL']
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40,30))
for i,app in enumerate(apps):
    df2 = most_insalled_apps[most_insalled_apps.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().sort_values(ascending=False).reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.

## How many apps are there on Google Play Store which get 5 ratings??
