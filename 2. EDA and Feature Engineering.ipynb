# Flight Price Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fp=pd.read_excel('flight_price.xlsx')
fp.head()
```

```python
fp.describe()
```

```python
fp.info()
```

```python
fp.head()
```

```python
# handling date_of_journey
fp['date']=fp['Date_of_Journey'].str.split('/').str[0]
fp['month']=fp['Date_of_Journey'].str.split('/').str[1]
fp['year']=fp['Date_of_Journey'].str.split('/').str[2]
```

```python
fp
```

```python
fp.info()
```

```python
fp['date']=fp['date'].astype(int)
fp['month']=fp['month'].astype(int)
fp['year']=fp['year'].astype(int)
```

```python
fp.info()
```

```python
fp.head()
```

```python
# droping date of journey which is not required further
fp.drop('Date_of_Journey',axis=1, inplace=True)
```

```python
fp.head()
```

```python
# now handling Arrival time

fp['Arrival_Time']=fp['Arrival_Time'].apply(lambda x:x.split(' ')[0])
fp.head()
```

```python
fp['Arrival_hour']=fp['Arrival_Time'].str.split(":").str[0]
fp['Arrival_min']=fp['Arrival_Time'].str.split(":").str[1]

fp.head()
```

```python
#changing datatype
fp['Arrival_hour']=fp['Arrival_hour'].astype(int)
fp['Arrival_min']=fp['Arrival_min'].astype(int)
```

```python
fp.info()
```

```python
fp.drop('Arrival_Time',axis=1, inplace=True)
fp.head()
```

```python
fp['Departure_hour']=fp['Dep_Time'].str.split(':').str[0]
fp['Departure_min']=fp['Dep_Time'].str.split(':').str[1]
```

```python
fp['Departure_hour']=fp['Departure_hour'].astype(int)
fp['Departure_min']=fp['Departure_min'].astype(int)
```

```python
fp.info()
```

```python
fp.drop("Dep_Time",axis=1,inplace=True)
fp.head()
```

```python
#checking for missing values
fp.isnull().sum()
```

```python
# worrking with categorical features

fp['Total_Stops'].unique()
```

```python
fp['Total_Stops'].mode()
```

```python
fp['Total_Stops']=fp['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, np.nan:1})
fp.head()
```

```python
fp.drop('Route',axis=1,inplace=True)
fp.head()
```

```python
# handling the Duration using regex pattern

fp[['duration_in_hr', 'duration_in_min']] = fp['Duration'].str.extract(r'(?:(\d+)h)?\s?(?:(\d+)m)?').fillna(0).astype(int)
fp.tail()
```

```python
fp['total_duration_in_min'] = fp['duration_in_hr'] * 60 + fp['duration_in_min']
fp.head()
```

```python
encoder.fit_transform(fp[['Airline','Source','Destination','Additional_Info']]).toarray()
```

```python
hotencoded=pd.DataFrame(encoder.fit_transform(fp[['Airline','Source','Destination','Additional_Info']]).toarray(),columns=encoder.get_feature_names_out())
hotencoded.head()
```

```python
fp.drop(['Airline','Source','Destination','Duration'],axis=1,inplace=True)
fp.head()
```
